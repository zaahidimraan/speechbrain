```python
def resample_audio(src_file, dst_file, target_sr):
    """Resamples an audio file to the specified target sample rate."""
    y, sr = librosa.load(src_file, sr=None)  # Load original audio without resampling
    y_resampled = librosa.resample(y, orig_sr=sr, target_sr=target_sr)
    sf.write(dst_file, y_resampled, target_sr, format="wav")  # Save in WAV format

def resample_folder(src_folder, dst_folder, target_sr):
    """Resamples all WAV files in a source folder and saves them in the destination folder."""
    os.makedirs(dst_folder, exist_ok=True)  # Create destination folder if it doesn't exist
    for filename in os.listdir(src_folder):
        if filename.endswith(".wav"):
            src_path = os.path.join(src_folder, filename)
            dst_path = os.path.join(dst_folder, filename)
            resample_audio(src_path, dst_path, target_sr)


# if os.path.exists('MiniLibriMix/train/mix_clean'):
#    resample_folder('MiniLibriMix/train/mix_clean', 'MiniLibriMix/train/mix_clean', 16000)
# else:
#   print("Folder not found")
```

```python
def display_spectrogram(file_path):
    """Displays the spectrogram of an audio file."""
    y, sr = librosa.load(file_path, sr=None)
    plt.figure(figsize=(12, 4))
    D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)  # Use np.max
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title(f'Spectrogram of {file_path}')
    plt.show()
# if os.path.exists('MiniLibriMix/train/mix_clean/100-121669-0026_718-129597-0003.wav'):
#   display_spectrogram('MiniLibriMix/train/mix_clean/100-121669-0026_718-129597-0003.wav')
# else:
#   print("File not found")
```

```python
def move_and_delete_folders(source_root_dir,source_folders, destination_folder):
    """Copies folders and their contents to a destination, then deletes the originals."""

    for folder_name in source_folders:
        source_path = os.path.join(source_root_dir, folder_name)
        destination_path = os.path.join(destination_folder, folder_name)

        if os.path.exists(source_path):
            try:
                shutil.copytree(source_path, destination_path)  # Copy the folder and its contents
                shutil.rmtree(source_path)  # Delete the original folder
                print(f"Moved '{folder_name}' to '{destination_folder}' and deleted original.")
            except Exception as e:
                print(f"Error processing '{folder_name}': {e}")
        else:
            print(f"Warning: Folder '{folder_name}' not found.")

# # Example usage
# source_folders = ["mix_clean"]
# destination_folder = "DNS-dataset/datasets_fullband/clean_fullband/french_speech/M-AILABS_Speech_Dataset/fr_FR_190hrs_16k/mix/sous_les_mers"
# source_root_dir='MiniLibriMix/train/'
# move_and_delete_folders(source_root_dir,source_folders, destination_folder)
```

```python
def keep_random_wav_files(root_folder, num_files_to_keep):
    """Searches for WAV files in a folder and subfolders, keeping only a specified number in each folder."""

    for dirpath, dirnames, filenames in os.walk(root_folder):
        wav_files = [f for f in filenames if f.endswith(".wav")]
        if len(wav_files) > num_files_to_keep:
            files_to_remove = random.sample(wav_files, len(wav_files) - num_files_to_keep)
            for file in files_to_remove:
                os.remove(os.path.join(dirpath, file))
                print(f"Removed: {os.path.join(dirpath, file)}")

# Example usage:
root_folder = "DNS-dataset"  # Replace with your root folder path
num_files_to_keep = 400

# keep_random_wav_files(root_folder, num_files_to_keep)
```

### Precoss the RIR_table_simple.csv to remove some errors

```python
def process_csv(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path, header=None)

    # Replace '/impulse' with an empty string in all columns
    df = df.replace('/simulated_rirs_16k', '', regex=True)
    df = df.replace('RIRS_NOISES/', '', regex=True)
    df = df.replace('datasets/', 'DNS-dataset/', regex=True)

    # Write the modified data back to the same CSV file
    df.to_csv(file_path, header=False, index=False)

def main():
    filename="RIR_table_simple.csv"
    if filename.endswith('.csv'):
        print(f"Processing: {filename}")
        process_csv(filename)

main()
```

```python
def copy_file(src, dst):
    """
    Copy a file from src to dst. If the file already exists at the destination, it will be replaced.
    """
    if os.path.isfile(src):
        # Create the destination directory if it does not exist
        if not os.path.exists(dst):
            print("Destination Path does not exist")
        else:
            # Construct the full path of the destination file
            dst_file = os.path.join(dst, os.path.basename(src))

            # Copy the file to the destination directory
            shutil.copy2(src, dst_file)
            print(f"Copied {src} to {dst_file}")
    else:
        print(f"The source file {src} does not exist.")
```

### zip the dataO folder to see the output

```python
def zip_folder(folder_path, output_path):
    """
    Create a zip file of the specified folder.

    Parameters:
    folder_path (str): The path of the folder to be zipped.
    output_path (str): The output path of the zip file (without .zip extension).
    """
    if os.path.isdir(folder_path):
        shutil.make_archive(output_path, 'zip', folder_path)
        print(f"Folder '{folder_path}' has been successfully zipped into '{output_path}.zip'")
    else:
        print(f"The folder '{folder_path}' does not exist.")

# Example usage
folder_to_zip = 'dataO'
zip_file_output = 'data'  # No .zip extension needed

zip_folder(folder_to_zip, zip_file_output)
```
